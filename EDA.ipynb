CONVERTING CATEGORICAL DATA INTO NUMERICAL

```python
label_encoder = preprocessing.LabelEncoder()
loan['Loan_history']= label_encoder.fit_transform(loan['Loan_history'])
loan['Purpose']= label_encoder.fit_transform(loan['Purpose'])
loan['Loan_history'].value_counts().plot(kind='pie',autopct='%.2f')
```
